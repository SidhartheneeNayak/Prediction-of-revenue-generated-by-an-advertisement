```python
raw_data = pd.read_csv('/content/Train_Data.csv')
test_data = pd.read_csv('/content/Test_Data.csv')
raw_data.head()
```

```python
df_cat = raw_data.select_dtypes(include="object").columns.values
df_num = raw_data.select_dtypes(exclude="object").columns.values
```

```python
raw_data['day'] = raw_data['date'].str.slice(start=0,stop=2).astype(int)
raw_data['month'] = raw_data['date'].str.slice(start=3,stop=5).astype(int)
raw_data['year'] = raw_data['date'].str.slice(start=6)
raw_data['ad_no'] = raw_data['ad'].str.slice(start=3).astype(int)
raw_data['ad_grp'] = raw_data['adgroup'].str.slice(start=8)
raw_data.head()
```

```python
df = raw_data.drop(['date','adgroup','ad','campaign'],axis=1)
df_cat = df.select_dtypes(include="object").columns.values
df_num = df.select_dtypes(exclude="object").columns.values
df_cat
```

```python
df_encoded=pd.get_dummies(df,columns=['ad_grp'])
df_encoded=pd.get_dummies(df_encoded,columns=['year'])
# label_encoder = preprocessing.LabelEncoder()
# df_encoded['year']= label_encoder.fit_transform(df_encoded['year'])
df_encoded['ctr'] = df_encoded['clicks']/df_encoded['impressions']
df_encoded['cpc'] = df_encoded['cost']/df_encoded['clicks']
df_encoded['cpa'] = df_encoded['cost']/df_encoded['conversions']
df_encoded['cpc'].fillna(value=df_encoded['cpc'].mean(), inplace=True)
df_encoded['ctr'].fillna(value=df_encoded['ctr'].mean(), inplace=True)
df_encoded['cpa'].fillna(value=0, inplace=True)
df_encoded.head()
```

```python
df_encoded['cpa'] = df_encoded['cpa'].replace(np.inf, 0.4)
```

```python
df_encoded.corr().style.background_gradient(cmap='GnBu')
```

```python
test_data['day'] = test_data['date'].str.slice(start=0,stop=2).astype(int)
test_data['month'] = test_data['date'].str.slice(start=3,stop=5).astype(int)
test_data['year'] = test_data['date'].str.slice(start=6)
test_data['ad_no'] = test_data['ad'].str.slice(start=3).astype(int)
test_data['ad_grp'] = test_data['adgroup'].str.slice(start=8)
test_data.head()
```

```python
df_test = test_data.drop(['date','adgroup','ad','campaign'],axis=1)
df_cat = df_test.select_dtypes(include="object").columns.values
df_num = df_test.select_dtypes(exclude="object").columns.values
df_cat
```

```python
df_encoded_test=pd.get_dummies(df_test,columns=['ad_grp'])
df_encoded_test=pd.get_dummies(df_encoded_test,columns=['year'])
#df_encoded_test['year']= label_encoder.transform(df_encoded_test['year'])
df_encoded_test['ctr'] = df_encoded_test['clicks']/df_encoded_test['impressions']
df_encoded_test['cpc'] = df_encoded_test['cost']/df_encoded_test['clicks']
df_encoded_test['cpa'] = df_encoded_test['cost']/df_encoded_test['conversions']
df_encoded_test['cpc'].fillna(value=df_encoded['cpc'].mean(), inplace=True)
df_encoded_test['ctr'].fillna(value=df_encoded['ctr'].mean(), inplace=True)
# df_encoded_test['cpa'].fillna(value=0, inplace=True)
# df_encoded_test['year_2020'] = df_encoded_test['year_2021']*0
df_encoded_test.head()
```

```python
df_encoded_test['cpa'] = df_encoded_test['cpa'].replace(np.inf, 0.23)
df_encoded_test[df_encoded_test['cpa'] == np.inf]
```

```python
threshold=3
avg = df_encoded['conversions'].mean()
st = df_encoded['conversions'].std()
df_new = df_encoded[((df_encoded['conversions']-avg)/st)<=3]
avg = df_encoded['impressions'].mean()
st = df_encoded['impressions'].std()
df_new2 = df_new[((df_new['impressions']-avg)/st)<=3]
avg = df_encoded['clicks'].mean()
st = df_encoded['clicks'].std()
df_new3 = df_new2[((df_new2['clicks']-avg)/st)<=3]
avg = df_encoded['cost'].mean()
st = df_encoded['cost'].std()
df_new4 = df_new3[((df_new3['cost']-avg)/st)<=4.5]
df_new4.shape
```

```python
df_new4.corr().style.background_gradient(cmap='GnBu')
```

```python
X = df_new4.drop(['revenue','month','ad_no','cpc','year_2020'],axis=1)
y = df_new4['revenue']
X_test = df_encoded_test.drop(['month','ad_no','cpc'],axis=1)
print(X.columns)
print(X_test.columns)
```

```python
X_test = X_test[['impressions', 'clicks', 'cost', 'conversions', 'day', 'ad_grp_1',
       'ad_grp_2', 'ad_grp_3', 'ad_grp_4', 'year_2021', 'ctr', 'cpa']]
print(X.columns)
print(X_test.columns)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.1,random_state=2021)
print("X_train:", X_train.shape) # X_train
print("X_val:",X_val.shape) # X_test
print("y_train:",y_train.shape) # y_train
print("y_val:",y_val.shape) # y_test
```

```python
xgb = XGBRegressor(n_estimators=93, max_depth=3, learning_rate=0.06, colsample_bynode=0.72, booster='dart', subsample=0.45, random_state=2024, verbosity=2)
xgb.fit(X,y)
results = cross_val_score(xgb,X,y,cv=6)
print(results)
print(results.mean())
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
def baseline_model():
	# create model
	model = Sequential()
	model.add(Dense(15, input_dim=11, kernel_initializer='normal', activation='relu'))
	#model.add(Dense(15, kernel_initializer='normal', activation='relu'))
	model.add(Dense(9, kernel_initializer='normal', activation='relu'))
	model.add(Dense(1, kernel_initializer='normal'))
	# Compile model
	model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
	return model
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)
```

```python
# evaluate model
estimator = KerasRegressor(build_fn=baseline_model, epochs=180, batch_size=9, verbose=0)
kfold = KFold(n_splits=5)
estimator.fit(X_train,y_train)
y_pred = estimator.predict(X_val)
print("Baseline: %.2f MSE" % mean_squared_error(y_pred,y_val))
```

```python
target2 = estimator.predict(X_test)
target2[target2<0]=0
target2 = np.around(target2,decimals=2)
res = pd.DataFrame(target2) #preditcions are nothing but the final predictions of your model on input features of your new unseen test data
res.index = X_test.index # its important for comparison. Here "test_new" is your new test dataset
res.columns = ["revenue"]
res.to_csv("/content/prediction_results2.csv", index = False)
```

```python
from keras.callbacks import EarlyStopping
nn_model = Sequential()
nn_model.add(Dense(24, kernel_initializer='normal',input_dim = X_train.shape[1], activation='relu'))

# The Hidden Layers :
nn_model.add(Dense(15, kernel_initializer='normal',activation='relu'))
nn_model.add(Dense(15, kernel_initializer='normal',activation='relu'))
#NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))

# The Output Layer :
nn_model.add(Dense(1, kernel_initializer='normal',activation='linear'))

# Compile the network :
nn_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['accuracy'])
nn_model.summary()
```

```python
from keras.callbacks import ModelCheckpoint
# checkpoint_name = 'Weights-{epoch:03d}--{val_loss:.5f}.hdf5' 
# checkpoint = ModelCheckpoint(checkpoint_name, monitor='val_loss', verbose = 1, save_best_only = True, mode ='auto')
# callbacks_list = [checkpoint]
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1,patience=80)
```

```python
history = nn_model.fit(X_train, y_train, validation_data = (X_val,y_val), batch_size=10, epochs=600, verbose=0, callbacks=[es])
# evaluate the model
_, train_acc = nn_model.evaluate(X_train, y_train, verbose=0)
_, test_acc = nn_model.evaluate(X_val, y_val, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
#nn_model.fit(X, y, epochs=500, batch_size=32, validation_split = 0.24, callbacks=callbacks_list)
```
